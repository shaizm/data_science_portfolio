### <font size=5 face='Comic Sans MS' color="#007707"><left><b>Data Wrangling</font>

```python
df = df.rename(columns={'m_Mining ':'m_Mining','f_Mining ':'f_Mining'})
```

```python
# Finding the total number of people in each community per profession
df_religion = df.groupby(df['religion']).sum()
```

```python
# Summing the rows and columns to get the total count
df_religion['Population'] = df_religion.sum(axis = 1)
df_religion.loc['Total'] = df_religion.sum(axis = 0)
```

```python
df_religion = df_religion.rename(index = {'others':  'T,J,Z,B'})
df_religion
```

* T,J,Z,B stands for Tribal,Jews,Zorastrian(parsi),Bahai

### <font size=5 face='Comic Sans MS' color="#007707"><left><b>Exploring Data [EDA]</font>

#### <font size=4 face='Comic Sans MS' color="#FFC300"><left><u><b>Above Data in Percentage (by Row)</u></font>
     - To finding the % of a particular community in each field.
     - To answer questions like; what % of Sikh community work in agriculture?

```python
# Finding % of male cultivators in each community (Percentage along the row)
a = df_religion['m_cultivator']/df_religion['Population']*100
a
```

```python
listing=df_religion.columns.to_list()
indexes = df_religion.index.to_list()
```

```python
# Creating a new DataFrame for percentage
df_test = pd.DataFrame(a,columns=[listing[0]])
```

```python
# Iterating for all columns
listing=df_religion.columns.to_list()

for i in listing[1:]:
    a = (df_religion[i]/df_religion['Population'])*100
    df_test[i]=a
df_test
```

#### <font size=4 face='Comic Sans MS' color="#FFC300"><left><u><b>Above Data in Percentage (by Column)</u></font>
     - To finding the % of male cultivators per community out of total male 
       cultivators(Percentage along the column)
     - To answer questions like; Which community provides to national food requirement?

```python
b = (df_religion['m_cultivator']/df_religion['m_cultivator'][-1])*100
b
```

```python
# Creating another DataFrame
df_test2 = pd.DataFrame(b,columns=[listing[0]])
```

```python
# Iterating for all columns
for i in listing[1:]:
    b = (df_religion[i]/df_religion[i][-1])*100
    df_test2[i]=b
df_test2
```
